```python
VertexLP = pulp.LpProblem("VertexProblem",pulp.LpMinimize)

Nodes=range(7)
Color=range(5)


x = pulp.LpVariable.dicts("x",(Nodes,Color),0,None,pulp.LpInteger)

y = pulp.LpVariable.dicts("y",(Color),0,None,pulp.LpInteger)

#Objective Function
VertexLP += sum(y[k] for k in Color)
#assigns every node a color
for j in Nodes:
  VertexLP+= sum (x[j][k]for k in Color)==1
#each color no more than once per edge
for k in Color:
  for e in Edges:
    VertexLP+= x[e[0]][k]+x[e[1]][k]<=1
#if a color is used in any node, y[k]=1
for k in Color:
  for j in Nodes:
    VertexLP+= (x[j][k])<=y[k]





print(VertexLP)
























#for v in VertexLP.variables():
 # print(v.name, "=", v.varValue)
#for v in VertexLP.variables():
    #if v.varValue > 0.1:
        #print(v.name, "=", v.varValue)
#SolutionArray=np.asarray([[x[mine][city].varValue for city in Cities] for mine in Mines])
#print(SolutionArray)
```

```python
#reads the data and saves it in a variable "data", which is a list of dictionaries
link = 'https://public.opendatasoft.com/explore/dataset/us-zip-code-latitude-and-longitude/download/?format=json&q=MI&refine.state=MI&timezone=America/New_York'
response=requests.get(link)
data = response.json()
```

```python
#returns the distance in km between two cities in the dataset using the haversine formula
#https://en.wikipedia.org/wiki/Haversine_formula
#code adapted from http://code.activestate.com/recipes/576779-calculating-distance-between-two-geographic-points/
def dist(i,j):
  latdiff=np.radians(data[i]['fields']['latitude']-data[j]['fields']['latitude'])
  londiff=np.radians(data[i]['fields']['longitude']-data[j]['fields']['longitude'])

  a = np.sin(latdiff/2)**2 + np.cos(np.radians(data[i]['fields']['latitude'])
                                     ) * np.cos(np.radians(data[j]['fields']['latitude'])) * np.sin(londiff/2)**2
  c = 2 * np.arctan2(np.sqrt(a),  np.sqrt(1-a))
  return 6371 * c
```

```python
RadioLP = pulp.LpProblem("RadioTowerProblem",pulp.LpMinimize)

Frequencies=range(20)


a = pulp.LpVariable.dicts("a",(zip,Frequencies),0,None,pulp.LpInteger)

b = pulp.LpVariable.dicts("b",(Frequencies),0,None,pulp.LpInteger)

#Objective Function
RadioLP += sum(b[f] for f in Frequencies)
#forces every Radio Tower to have exactly one Frequency
for z in zip:
  RadioLP+= sum (a[z][f]for f in Frequencies)==1

#if a Frequency is used for any radio tower, b[f]=1
for f in Frequencies:
  for z in zip:
    RadioLP+= (a[z][f])<=b[f]































```

```python
for f in Frequencies:
  for c in Connect:
    RadioLP+= a[c[0]][f]+a[c[1]][f]<=1
print (RadioLP)
#This Radio Tower ensures that no frequency is used more than once for any two towers within 70 miles.
```

```python
RadioLP.solve()
print("Status:", pulp.LpStatus[RadioLP.status])
print("Result:", (pulp.value(RadioLP.objective)))
for v in RadioLP.variables():
    if v.varValue > 0.1:
        print(v.name, "=", v.varValue)
```
